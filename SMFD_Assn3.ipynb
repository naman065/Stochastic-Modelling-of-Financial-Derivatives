```python

import numpy as np
import matplotlib.pyplot as plt

# Parameters
T = 1.0             # Total time
N = 1000            # Number of steps
dt = T / N          # Time step size
t = np.linspace(0, T, N+1)  # Time grid

# Wiener process: W(0) = 0, increments ~ N(0, dt) (normal distribution)
np.random.seed(42)
dW = np.random.normal(0, np.sqrt(dt), size=N)  # Increments
W = np.zeros(N+1)   # Initializing Wiener path
W[1:] = np.cumsum(dW)  # Constructing the Brownian path

# Plotting the path
plt.figure(figsize=(10, 5))
plt.plot(t, W, label='Wiener Process Path', color='blue')
plt.title('Simulation of 1D Standard Brownian Motion')
plt.xlabel('Time t')
plt.ylabel('W(t)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
S0 = 100           # Initial stock price
T = 1.0            # Time horizon
N = 1000           # Number of time steps
dt = T / N         # Time step size
t = np.linspace(0, T, N + 1)

alpha = 0.1        # COnstants
sigma = 0.2
num_paths = 5      # Number of GBM paths to simulate


np.random.seed(0)

S_paths = np.zeros((num_paths, N + 1))
S_paths[:, 0] = S0

for i in range(num_paths):
    W = np.cumsum(np.random.normal(0, np.sqrt(dt), size=N))  # Brownian increments
    W = np.insert(W, 0, 0)  #  W(0) = 0
    exponent = sigma * W + (alpha - 0.5 * sigma**2) * t
    S_paths[i] = S0 * np.exp(exponent)


plt.figure(figsize=(10, 6))
for i in range(num_paths):
    plt.plot(t, S_paths[i], label=f"Path {i+1}")

plt.title("5 Sample Paths of Geometric Brownian Motion")
plt.xlabel("Time t")
plt.ylabel("Stock Price S(t)")
plt.grid(True)
plt.legend()
plt.show()
```
